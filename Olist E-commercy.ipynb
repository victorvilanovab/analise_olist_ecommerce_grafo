```python
import pandas as pd
```

```python
df_costumers = pd.read_csv("data_compras/olist_customers_dataset.csv")
df_costumers.head()
```

```python
df_sellers = pd.read_csv("data_compras/olist_sellers_dataset.csv")
df_sellers.head()
```

```python
df_orders = pd.read_csv("data_compras/olist_orders_dataset.csv")
df_orders.head()
```

```python
df_orders_sellers = pd.read_csv("data_compras/olist_order_items_dataset.csv")
df_orders_sellers.head()
```

```python
merge_df = pd.merge(df_orders_sellers, df_orders, on = "order_id")
merge_df = pd.merge(merge_df, df_sellers , on = "seller_id")
merge_df = pd.merge(merge_df, df_costumers , on = "customer_id")
merge_df
```

```python
graphi_table = merge_df[["customer_city", "customer_state", "seller_city", "seller_city"]]
graphi_table
```

```python
graphi_table.dropna()
```

```python
lista = [1 for i in range(len(graphi_table))]
```

```python
graphi_table["peso"] = lista
```

```python
graphi_table = graphi_table.loc[:,~graphi_table.columns.duplicated()]
graphi_table
```

```python
df_geolocation = pd.read_csv("data_compras/olist_geolocation_dataset.csv")
df_geolocation = df_geolocation[["geolocation_city", "geolocation_state"]]
df_geolocation = df_geolocation.drop_duplicates().reset_index(drop=True)
df_geolocation
```

```python
df_geolocation.geolocation_city
```

```python
#creating dict to map
maping = {}
for i in range(len(df_geolocation)):
    maping[df_geolocation.geolocation_city[i]] = df_geolocation.geolocation_state[i]
```

```python
graphi_table
```

```python
graphi_table["seller_state"] = graphi_table["seller_city"].map(maping)
```

```python
graphi_table
```

```python
graphi_table_export = graphi_table[[ "seller_state","customer_state", "peso"]]
graphi_table_export.head()
```

```python
arestas = graphi_table_export.rename(columns = {"seller_state": "Source","customer_state": "Target", "peso": "Weight"})
arestas.head()
```

```python
arestas = arestas.groupby(['Source','Target'])['Weight'].sum().reset_index()
arestas.head()
```

```python
arestas.sort_values(by = "Weight")
```

```python
nodes = df_geolocation["geolocation_state"]
nodes = pd.DataFrame(nodes)
nodes = nodes.drop_duplicates()
lista_index = [ i for i in range(len(nodes))]
nodes["Id"] = lista_index
nodes = nodes.rename(columns = {"geolocation_state": "Label"})
nodes.reset_index(inplace = True, drop = True)
nodes.head()
```

```python
maping2 = {}
for i in range(len(nodes)):
    maping2[nodes.Label[i]] = nodes.Id[i]
```

```python
maping2
```

```python
arestas.Source = arestas.Source.map(maping2)
arestas.Target = arestas.Target.map(maping2)
arestas.head()
```

```python
nodes.to_csv("nodes.csv",index = False)
arestas.to_csv("arestas.csv", index = False)
```
